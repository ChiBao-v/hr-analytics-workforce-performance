## 1. Import Library

```python
import pandas as pd
import sqlite3 
from pathlib import Path 
```

## 2. Connect to SQLite database

```python
BASE_DIR = Path.cwd().parent
RAW_DIR = BASE_DIR / 'data' / 'raw'
DB_PATH = BASE_DIR / 'sql' /'hr_analytics.db'

print("BASE_DIR:", BASE_DIR)
print("RAW_DIR:", RAW_DIR)
print("DB_PATH:", DB_PATH)
```

```python
conn = sqlite3.connect(DB_PATH)

# CSV file to load
csv_files = {
    'employees': RAW_DIR / 'employees.csv',
    'stores': RAW_DIR / 'stores.csv',
    'monthly_performance': RAW_DIR / 'monthly_performance.csv',
    'role_kpis': RAW_DIR / 'role_kpis.csv',
    'business_outcomes': RAW_DIR / 'business_outcomes.csv'
}

# Load CSV file into SQLite
for table_name, file_path in csv_files.items():
    df = pd.read_csv(file_path)
    df.to_sql(table_name, conn, if_exists='replace', index=False)
    print(f'Loaded {table_name}: {df.shape[0]} rows, {df.shape[1]} columns')

conn.close

print(f'Database created at: {DB_PATH}')
```
